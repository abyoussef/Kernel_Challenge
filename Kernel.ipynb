```python
def plot_img( l ) : 
    size_img  = 32 * 32 
    image_shape = (32,32)
    rgb_shape = (32,32,3) 
    for i in range(3): 
        x = l[i*size_img :(i+1)*size_img]
        channel = inv_preproc(x).reshape(image_shape)
        rgb_array[:,:,i] = np.array(channel )
    plt.imshow(rgb_array)
    plt.show()
    return
```

```python
def inv_preproc(l) : 
    x = l / np.sqrt(np.var(l)) 
    x_ = x - np.min(x)
    x__ = x_ / (np.max(x_) - np.min(x_)) 
    x_im = x__ * 254 +1
    return x_im
```
